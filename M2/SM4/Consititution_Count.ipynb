```python
#Entrypoint 2.x
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Spark SQL basic example").enableHiveSupport().getOrCreate()

# On yarn:
# spark = SparkSession.builder.appName("Spark SQL basic example").enableHiveSupport().master("yarn").getOrCreate()
# specify .master("yarn")

sc = spark.sparkContext
```

```python
file_path = "file:///home/talentum/test-jupyter/P2/M2/SM4/4_AdvancedRddActions/constitution.txt"

# Create a baseRDD from the file path
baseRDD1 = sc.textFile(file_path)


# Split the lines of baseRDD into words
splitRDD1 = baseRDD1.flatMap(lambda x: x.split(" "))


# Create a tuple of the word and 1 
splitRDD_no = splitRDD1.map(lambda w: (w, 1))

# Count of the number of occurences of each word
resultRDD1 = splitRDD_no.reduceByKey(lambda x, y: x + y)

#swap
rddSwap = resultRDD1.map(lambda t : (t[1],t[0]))

#sort
rddSort = rddSwap.sortByKey(ascending = False)

rddSort.take(5)
```
